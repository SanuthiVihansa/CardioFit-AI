```python
# Load your video or sequence of images
cap = cv2.VideoCapture(vidpath)

# Parameters for Farneback optical flow
farneback_params = dict(pyr_scale=0.5, levels=3, winsize=15, iterations=3, poly_n=5, poly_sigma=1.2, flags=0)

# Take first frame and convert it to gray
ret, old_frame = cap.read()
old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # print the time of the video in seconds using a condition so that it prints only every second the video has completed reading
    # if cap.get(cv2.CAP_PROP_POS_MSEC) == 1000:
    print("Seconds: ", cap.get(cv2.CAP_PROP_POS_MSEC) / 1000)


    # stop loop after 10 seconds   
    if cap.get(cv2.CAP_PROP_POS_MSEC) > 10000:
        break

    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Calculate optical flow
    flow = cv2.calcOpticalFlowFarneback(old_gray, frame_gray, None, **farneback_params)

    # Now update the previous frame and previous points
    old_gray = frame_gray.copy()

cv2.destroyAllWindows()
cap.release()
```
